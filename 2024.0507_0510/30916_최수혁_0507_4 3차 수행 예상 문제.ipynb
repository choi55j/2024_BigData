## 수행 평가 예상 문제

```python
# 빈 딕셔너리 만들기
dict_word = {}

# 공백을 입력할때 까지 입렵 키값과 밸류값 받기
while True:
    input_word=input('영단어, 뜻: ')
    if input_word =='':
        break
    # 키 값과 밸류 값 분리하기
    eng = input_word.split(',')[0]
    kor = input_word.split(',')[1]
    
    # 키 값과 밸류 값 저장하기
    dict_word[eng] = kor

# 출력하기
dict_word
```

```python
price = int(input('물건 구매가 : '))

# 할인율 결정
if price >= 100000:
    dc = 10
elif price >= 50000:
    dc = 7
elif price >= 10000:
    dc = 5
else:
    dc = 0
    
# 출력
print(f'''
구매가 : {price}
할인율 : {dc}%
할인금액 : {price*(dc/100)}
지불금액 : {price - price*(dc/100)}
''')
```

```python
num1 = int(input('숫자 1 : '))
num2 = int(input('숫자 2 : '))
op = input('연산 기호 : ')

if op == '+':
    print(f'덧셈 : {num1} + {num2} = {num1+num2}')
elif op == '-':
    print(f'뺄셈 : {num1} - {num2} = {num1-num2}')
elif op == '*':
    print(f'곱셈 : {num1} * {num2} = {num1*num2}')
elif op == '/':
    print(f'나눗셈 : {num1} / {num2} = {num1/num2}')
else:
    print('잘못 입력하셨습니다!')
```

```python
import random
def guess_number_game(max_attempts=5):
    cnt = 0
    num= random.randint(1,100)
    
    while cnt < max_attempts:
        cnt +=1
        answer = int(input(f'Attempt {cnt}. 예상 숫자를 입력하세여 (1-100) : '))
        if answer == num:
            print('정답입니다!')
            print(f'{cnt}번 만에 맞추었습니다.')
            return
        if answer < num:
            print('Up')
        else:
            print('Down')
    print(f'횟수 초과입니다. 정답은 {num}였습니다')
# 시작
guess_number_game()
```

```python
# 주어진 아이디와 비밀번호
id = 'id123'
pwd = 'pwd123'

while True:
    input_id = input('id : ')
    input_pwd = input('pwd : ')
    
    if id == input_id and pwd == input_pwd:
        break
    if id != input_id:
        print('아이디를 확인하세요')
    if pwd != input_pwd:
        print('비밀번호를 확인하세요')
```
