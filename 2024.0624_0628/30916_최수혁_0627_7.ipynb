```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

plt.figure(dpi=100)

plt.title('월별 판매 실적')

plt.plot(month,sales, color='m',label='단위:개수')

plt.legend(loc='lower right')

plt.xlabel('월')
plt.ylabel('매출')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

import random

he=[]
we=[]
for i in range(100):
    he.append(random.randint(100,200))
    we.append(random.randint(20,100))
    
plt.title('키와 몸무게의 상관관계')

plt.scatter(he,we,c=he,cmap='RdPu')

plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

import random

he,we,sc = [],[],[]
for i in range(100):
    he.append(random.randint(100,200))
    we.append(random.randint(20,100))
    sc.append(random.randint(1,1000))
    
plt.title('키, 몸무게 상관')
plt.scatter(he,we,s=sc,c=sc,cmap='rainbow',alpha=0.7)

plt.colorbar(label='체육 실기 점수')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [41,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i]
    
plt.rcParams['axes.unicode_minus']=False

plt.title('오디션 득표 현황')

plt.barh(singer,week1,label='첫째주',color='hotpink')
plt.barh(singer,week2,label='둘째주',color='royalblue')

plt.legend()
plt.show()
```
