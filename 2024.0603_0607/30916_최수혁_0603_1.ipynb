```python
# 꺽은선 그래프 그리기
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글 폰트 적용하기
plt.rc('font',family='Malgun Gothic')

# x,y출 데이터 지정
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 그래프의 해상도 설정하기 dpi / figure, 1인치 안에 점이 150개
plt.figure(dpi=150)

# 제목 title
plt.title('월별 판매 실적')

# 꺽은 선형 그래프 그리기 .plot /.color
# 그래프 범위 .label
plt.plot(month,sales,color='y', label='단위계수')
# 범례를 보이게 설정,loc =  lower right
plt.legend(loc = 'lower right')
# x축, y축 / xlabel, ylabel
plt.xlabel("월")
plt.ylabel('매출')

# 그래프 보이기
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

sales = [1,5,7,3,7]
month = [3,4,5,6,7]
plt.figure(dpi=150)
# 막대 그래프 그리기 plt.bar
plt.title('월별 판매 실적')
plt.bar(month,sales, color='b')

plt.show()
```
